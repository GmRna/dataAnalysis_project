```python
# 아파트 비율, 거주인구비율, 직장인구 비율, 1인가구 비율, 65세이상인구 비율, 20대인구 비율
col_list = []
row_list = []
for i in range(0, len(dong)):
    url = "https://sgisapi.kostat.go.kr/OpenAPI3/startupbiz/regiontotal.json"

    params = {
        "accessToken": "{}".format(token),
        "adm_cd": "{}".format(dong['addr_code'][i])  
    }

    response = requests.get(url, params=params)
    json = response.json()
    if i == 0:
        col_list = list(json['result'][1].keys())
    row_list.append(json['result'][1].values())
df = pd.DataFrame(row_list, columns=col_list)
```
