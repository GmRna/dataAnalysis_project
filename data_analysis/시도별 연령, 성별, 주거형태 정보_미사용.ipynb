### 서울시 시도별 연령, 성별, 주거형태 정보
 - 사용하려했으나 따로 잘정리된 API 존재하여서 제외 후 해당 데이터 사용

```python
import pandas as pd
import numpy as np
import requests, json
```

```python
dong = pd.read_csv('seoul_dong.csv')
dong
```

```python
# 서울시 동별 인구, 연령정보 API
addr_len = len(dong['addr_code'])
rowList = []
columnList = []
flag = True
for i in range(0,addr_len):
    # API 호출
    url = "https://sgisapi.kostat.go.kr/OpenAPI3/startupbiz/pplsummary.json"

    params = {
        "accessToken": "bec0b6ea-0612-430a-8e4a-d5ecb04db93a",
        "adm_cd": "{}".format(dong['addr_code'][i]) # 동별 인구데이터 호출
    }

    response = requests.get(url, params=params)
    json = response.json()
    # csv를 만들기 위한 값 저장
    resultLen = len(list(json['result'][0].keys())) # 호출받은 데이터 길이
    if flag:
        for c in range(0, resultLen):
            columnList.append(list(json['result'][0].keys())[c]) # 컬럼값 저장
            flag = False
    loc = []
    for col in range(0, len(columnList)):
        loc.append(json['result'][0].get('{}'.format(list(json['result'][0].keys())[col]))) # 로우값 저장
    print(loc)
    rowList.append(loc)
```

```python
# csv 파일 저장
result = pd.DataFrame(rowList, columns=columnList)
result.to_csv('seoul_population.csv', index=False)
```
